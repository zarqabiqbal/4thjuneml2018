```python
import cv2

#load local image
img1=cv2.imread("Cat.jpg")
#initialize video capture device
capture=cv2.VideoCapture(0)
#check whether camera is opened or not
while capture.isOpened():
    print("Camera is Starting ........")
    #capture the image from the camera and check its status
    status,img2=capture.read()
    #crop the part from the camera image
    part_of_image=img2[100:200,300:400]
    #add the crop part in local image
    img1[100:200,300:400]=part_of_image
    #show the image
    cv2.imshow("Local Image",img1)
    cv2.imshow("Camera Image",img2)
    #wait for the q key to quit the window
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
#after pressing any key it destroy all window which open
cv2.destroyAllWindows()
#release the camera
capture.release()
```
